# Theory

$$\mid<\bar{P}^{0}|P(t)>\mid ^{2} = \mid g_{-}(t) \mid ^{2} \mid \frac{p}{q} \mid^{2}$$
$$\mid<{P}^{0}|P(t)>\mid ^{2} = \mid g_{+}(t) \mid ^{2}$$


with $ \mid g_{\pm}(t) \mid^{2} = \frac{e^{-\Gamma t}}{2} (\cosh{\frac{\Delta \Gamma t}{2}} \pm \cos{\frac{\Delta m t}{2}})$

# parameters value
Values of $\Delta \Gamma$ and $\Delta m$ for $K, D, B, B_{s}$ systems.

Define: \\

$$x= \frac{\Delta m}{\Gamma}$$,

$$y= \frac{\Delta \Gamma}{2\Gamma}$$ and
$$\tau = 1/\Gamma$$

So, eventually the formulae of $g_{\pm}$,

$$ \mid g_{\pm}(t) \mid^{2} = \frac{e^{- t/\tau}}{2} (\cosh{\frac{y t}{\tau}} \pm \cos{\frac{x t}{\tau}})$$


meson system | $\tau$ | $x$ | $y$
----------|---------|----------|---------
$K$ system   | 0.26 ns$^{-1}$ |  0.477 | -1|
$D$ system   | 0.41 ps$^{-1}$ |  0.0046 | 0.0062|
$B$ system   | 1.52 ps$^{-1}$ |  0.769 | 0.0005|
$B_{s}$ system   | 1.53 ps$^{-1}$ |  26.8 | 0.068|

```python
import math
import numpy as np
import matplotlib.pyplot as plt

def gplus(tau, x, y, t, plus):
    if plus:
        Gp = math.e**(-t/tau) * 0.5 * (np.cosh(y*t/tau) + (np.cos(x*t/tau)))
        return Gp
    else:
        Gm =math.e**(-t/tau) * 0.5 * (np.cosh(y*t/tau) - (np.cos(x*t/tau)))
        return Gm
```

# Plotting the probablility

## kaon system

```python
tau = 0.26
x = 0.477
y =  -1
t = np.linspace(0,2,100)
y1 = gplus(tau, x, y, t, plus=True)
y2 = gplus(tau, x, y, t, plus=False)


plt.plot(t,y2, label = r'$P(K^{0}(t) \to \bar{K^{0}}(t)$', linestyle='--', color= 'r')
plt.plot(t,y1, label = r'$P(K^{0}(t) \to {K^{0}(t)}$')
plt.xlabel('t [ns]')
plt.ylabel('Probability')
plt.legend()
plt.show()
```
